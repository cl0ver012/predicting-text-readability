# Formula analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from scipy.stats import spearmanr
```

Get the WeeBit dataset with features.

```python
df = pd.read_csv("../features/weebit_train_with_features.csv", index_col=0)
```

## 1. Flesch reading ease

```python
from readability_formulas import flesch
df = flesch(df)
```

```python
print(spearmanr(df['Level'], df['Flesch']))
```

```python
plt.figure(figsize=(10, 8))
sns.boxplot(x="Level", y="Flesch", data=df)
plt.title("Distribution of Flesch score per readability level")
plt.show()
```

## Dale-Chall readability formula

```python
from readability_formulas import dale_chall
df = dale_chall(df)
```

```python
print(spearmanr(df['Level'], df['Dale_Chall']))
```

```python
plt.figure(figsize=(10, 8))
sns.boxplot(x="Level", y="Dale_Chall", data=df)
plt.title("Distribution of Dale-Chall score per readability level")
plt.show()
```

```python
from readability_formulas import gunning_fog
df = gunning_fog(df)
```
